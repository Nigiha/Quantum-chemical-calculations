### s型GTOについて
s型GTOは次の形で書ける。
$$
\begin{align*}
G_{000}(\bf{r}, \alpha, \bf{A})&=\exp(-\alpha|\bf{r}-\bf{A}|^2) \\
&=\exp[-\alpha\{(x-A_x)^2+(y-A_y)^2+(z-A_z)^2\}] \\
\end{align*}
$$
Gについている下付き文字はxyz成分の角運動量を表す。ここではs型に限定しているので$(l_x, l_y, l_z)=(0, 0, 0)$である。x, y, z成分に分解すると、
$$
\begin{align*}
G_{000}(\bf{r}, \alpha, \bf{A})&=\exp[-\alpha(x-A_x)^2]\ \cdot\ \exp[-\alpha(y-A_y)^2]\ \cdot \ \exp[-\alpha(z-A_z)^2] \\
&=G_0(x, \alpha, A_x)\ \cdot \ G_0(y, \alpha, A_y)\ \cdot \ G_0(z, \alpha, A_z)
\end{align*}
$$
よって、GTOの積分計算をするときにはxyzのそれぞれの積分に分解して考えられることが分かる。
以降、x成分のみを考えることとする。
ここで、s型GTOの規格化定数$N_{\alpha}$について考えておく。$N_{\alpha}$が満たすべき条件は次式である。
$$
\begin{align*}
N_{\alpha}^2 \int^{\infty}_{-\infty}G_0^2(x, \alpha, A_x)dx&=1 \\
N_{\alpha}^2 \int \exp^2[-\alpha(x-A_x)^2]dx&=1 \\
N_{\alpha} &=(\frac{2\alpha}{\pi})^{\frac{1}{4}}
\end{align*}
$$


```python
def N_a(a:float):
    return np.power(2*a/np.pi, 1/4)

print(N_a(1))
```

### 重なり積分
s型GTO同士の重なり積分は次の形で表せる。
$$
\begin{align*}
S_{\alpha\beta}&=N_{\alpha}N_{\beta}\int G_0(x, \alpha, A_x)G_0(x, \beta, B_x)dx \\
&=N_{\alpha}N_{\beta}\int \exp[-\alpha(x-A_x)^2]\exp[-\beta(x-B_x)^2]dx
\end{align*}
$$
教科書の式(2.7)の多中心積分の公式を使えば、
$$
\begin{align*}
S_{\alpha\beta}&=N_{\alpha}N_{\beta}K_{\alpha\beta, x}\int \exp(-px_p^2) dx\\
&=N_{\alpha}N_{\beta}K_{\alpha\beta, x}\sqrt{\frac{\pi}{p}}
\end{align*}
$$

```python
def overlap_int(a:float, b:float, A_x:float, B_x:float):
    p=a+b
    q=a*b/p
    X_AB=abs(A_x-B_x)
    K=np.exp(-q*X_AB**2)

    return N_a(a)*N_a(b)*K*np.sqrt(np.pi/p)

print()
```
