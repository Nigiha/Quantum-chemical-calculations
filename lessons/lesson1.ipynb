# Hartree Fockの実装(1)　-重なり積分と運動エネルギー積分-

### s型GTOについて
s型GTOは次の形で書ける。
$$
\begin{align*}
G_{000}(\bf{r}, \alpha, \bf{A})&=\exp(-\alpha|\bf{r}-\bf{A}|^2) \\
&=\exp[-\alpha\{(x-A_x)^2+(y-A_y)^2+(z-A_z)^2\}] \\
\end{align*}
$$
Gについている下付き文字はxyz成分の角運動量を表す。ここではs型に限定しているので$(l_x, l_y, l_z)=(0, 0, 0)$である。\
規格化定数$N_{\alpha}$について考えておく。$N_{\alpha}$が満たすべき条件は次式である。
$$
\begin{align*}
N_{\alpha}^2 \int^{\infty}_{-\infty}G_0^2(x, \alpha, A_x)dx&=1 \\
N_{\alpha}^2 \int \exp^2[-\alpha(x-A_x)^2]dx&=1 \\
N_{\alpha} &=(\frac{2\alpha}{\pi})^{\frac{1}{4}}
\end{align*}
$$


```python
#規格化定数
def N_a(a:float):
    return np.power(2*a/np.pi, 1/4)

print(N_a(1))
```

### 重なり積分
s型GTO同士の重なり積分は次の形で表せる。
$$
\begin{align*}
S_{\alpha\beta}&=N_{\alpha}N_{\beta}\int G_0(\bf{r}, \alpha, \bf{A})G_0(\bf{r}, \beta, \bf{B})d\bf{r} \\
&=N_{\alpha}N_{\beta}\int \exp[-\alpha(\bf{r}-\bf{A})^2]\exp[-\beta(\bf{r}-\bf{B})^2]d\bf{r}
\end{align*}
$$
教科書の式(2.17)より、
$$
\begin{align*}
S_{\alpha\beta}&=\left(\frac{4\mu}{p}\right)^{3/4}K_{\alpha\beta}
\end{align*}
$$
ここで、
$$
\begin{align*}
p=\alpha+\beta, \ \mu=\alpha\beta/p, \ \bf{R}=\bf{A}-\bf{B}, \ K=\exp(-\mu\bf{R}^2)
\end{align*}
$$
とした。

```python
'''
a: orbital exponent on Gaussian 'a' (e.g. alpha in the text)
b: orbital exponent on Gaussian 'b' (e.g. beta in the text)
A: list containing origin of Gaussian 'a' (e.g. [1.0, 2.0, 0.0])
B: list containing origin of Gaussian 'b'
'''
    
def p(a:float, b:float):
    return a+b

def mu(a:float, b:float):
    return a*b/p(a, b)

def R(A:list, B:list):
    return np.linalg.norm(A, B) 

def K(a:float, b:float, A:np.array, B:np.array):
    return np.exp(-mu(a, b)*R(A, B)**2)


#重なり積分
def overlap_interal(a:float, b:float, A:np.array, B:np.array):
    
    p=p(a, b)
    mu=mu(a, b)
    K=K(a, b, A, B)
    return np.power((4*mu/p), 3/4)*K

print()
```

### 運動エネルギー積分
運動エネルギー積分は次の形で表せる。

$$
\begin{align*}
T_{\alpha\beta}&=N_{\alpha}N_{\beta}\int G_{000}(\bf{r}, \alpha, \bf{A}) (-\frac{1}{2}\nabla^2) G_{000}(\bf{r}, \beta, \bf{B})d\bf{r}
\end{align*}
$$
教科書の式(2.19)より、
$$
\begin{align*}
T_{\alpha\beta}&=\mu(3-2\mu R
_{AB}^2)\left(\frac{4\mu}{p}\right)^{3/4}K_{\alpha\beta}
\end{align*}
$$


```python
#運動エネルギー積分
def kinetic_integral(a:float, b:float, A:np.array, B:np.array):
    '''
    a: orbital exponent on Gaussian 'a' (e.g. alpha in the text)
    b: orbital exponent on Gaussian 'b' (e.g. beta in the text)
    A: list containing origin of Gaussian 'a' (e.g. [1.0, 2.0, 0.0])
    B: list containing origin of Gaussian 'b'
    '''
    p=p(a, b)
    mu=mu(a, b)
    R=R(A, B)
    K=K(a, b, A, B)

    return mu*(3-2*mu*R**2)*np.power((4*mu/p, 3/4))
```
