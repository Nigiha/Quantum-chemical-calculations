```python
'''
a: orbital exponent on Gaussian 'a' (e.g. alpha in the text)
b: orbital exponent on Gaussian 'b' (e.g. beta in the text)
A: list containing origin of Gaussian 'a' (e.g. [1.0, 2.0, 0.0])
B: list containing origin of Gaussian 'b'
'''


#規格化定数
def N_a(a:float):
    return np.power(2*a/np.pi, 1/4)

print(N_a(1))


#文字の定義
def p(a:float, b:float):
    return a+b

def mu(a:float, b:float):
    return a*b/p(a, b)

def R(A:list, B:list):
    return np.linalg.norm(A, B) 

def K(a:float, b:float, A:np.array, B:np.array):
    return np.exp(-mu(a, b)*R(A, B)**2)


#重なり積分
def overlap_interal(a:float, b:float, A:np.array, B:np.array):
    
    p=p(a, b)
    mu=mu(a, b)
    K=K(a, b, A, B)
    return np.power((4*mu/p), 3/4)*K


#運動エネルギー積分
def kinetic_integral(a:float, b:float, A:np.array, B:np.array):

    p=p(a, b)
    mu=mu(a, b)
    R=R(A, B)
    K=K(a, b, A, B)

    return mu*(3-2*mu*R**2)*np.power((4*mu/p, 3/4))
```
